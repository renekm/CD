### Itens 4 a 6 - Resolução:

## O histograma conseguido pode ser analisado e dito que se parece com os seguintes tipos de distribuições teoricas: f e betaprime. Isso significa que o histograma da um pico no inicio e cai bruscamente e no final ele diminui lentamente. 

```python
dfn = 25
dfd = 15
x = np.linspace(f.ppf(0.01, dfn, dfd),f.ppf(0.999, dfn, dfd), 100)
```

```python
 y = f.pdf(x,dfn,dfd)
```

```python
histograma.plot(kind='hist', legend=False, bins = 150, normed = True, color = 'black')
plt.plot(x,y, color = 'r')
```

```python
k = 10.4
s = 3.6
z = np.linspace(mielke.ppf(0.01, k, s), mielke.ppf(0.999, k, s), 100)
```

```python
w = mielke.pdf(x,k, s)
```

```python
histograma.plot(kind='hist', legend=False, bins = 150, normed = True, color ='black')
plt.plot(z,w, color = 'green')
```

```python
histograma.plot(kind='hist', legend=False, bins = 150, normed = True, color ='black')
plt.plot(z,w, color = 'green')
plt.plot(x,y, color = 'red')

plt.show()
```

### Itens 7 a 9 - Resolução:

```python
fra = []
for i in range (len(dados_regiao_pib)):
    acc = (i - 0.5) / len(dados_regiao_pib)
    if acc < 0:
        acc = 0
    fra.append(acc) 
    
fra = pd.Series(data=fra)

q_f = stats.f.ppf(fra,dfn, dfd)
q_mielke = stats.mielke.ppf(fra,k,s)

fig = plt.figure(figsize=(12,8))

f_plot = plt.plot(dados_regiao_pib, q_f, 'o', color="red", label="f")
mielke_plot = plt.plot(dados_regiao_pib, q_mielke, 'o', color="green", label = "Mielke")

plt.legend(handles=[f_plot, mielke_plot])
plt.plot((min(dados_regiao_pib), max(dados_regiao_pib)), (min(dados_regiao_pib), max(dados_regiao_pib)), '-')

for i in range(len(dados_regiao_pib)):
    plt.plot((dados_regiao_pib[i], dados_regiao_pib[i]), (q_f[i], dados_regiao_pib[i]), '-', color='red', alpha = 0.3 )
    plt.plot((dados_regiao_pib[i], dados_regiao_pib[i]), (q_mielke[i], dados_regiao_pib[i]), '-', color='green', alpha = 0.3)
```
